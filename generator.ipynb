```python
#Define a generator function prime_number_generator that yields prime numbers indefinitely. Print the first 15 prime numbers generated by this function.

def prime_number_generator():
    num = 2
    while True:
        for i in range(2, int(num ** 0.5) + 1):
            if num % i == 0:
                break
        else:
            yield num
        num += 1


gen = prime_number_generator()
for _ in range(15):
    print(next(gen), end=' ')
```

```python
#mplement a generator function fibonacci_generator that yields Fibonacci numbers indefinitely. Write a function sum_fibonacci that computes the 
#sum of the first 10 Fibonacci numbers generated by this function.

def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

def sum_fibonacci(n):
    gen = fibonacci_generator()
    total = 0
    for _ in range(n):
        total += next(gen)
    return total

print("Sum of the first 10 Fibonacci numbers:", sum_fibonacci(10))
```
